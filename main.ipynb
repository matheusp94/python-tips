if else trick


```python
#instead of

'''a = 20
b = 10
if (a > b):
    print("o maior é ", a)
else:
    print("o maior é ", b)'''

#usar

a = 20
b = 10
res = a if a > b else b
print("O maior valor é ", res)
```

stockchart plot

```python
import yfinance as yf
import matplotlib.pyplot as plt

# download historical data os stock.ns
ticker = input('Entre com o nome da ação: ')

try:
    data = yf.download(ticker, start='2021-01-01', end='2023-10-15')

    # plot das ações
    plt.figure(figsize=(10, 5))
    plt.plot(data['Close'])
    plt.title(f'{ticker} stock chart')
    plt.xlabel('Data')
    plt.ylabel('Preço (INR)')
    plt.show()

except Exception as e:
    print(f'Erro: {e}')
```
